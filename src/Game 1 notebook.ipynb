```python
from highlight_utils import create_highlight_clip
from moviepy.editor import VideoFileClip, CompositeVideoClip, TextClip


# === Load your two camera feeds ===
video0 = VideoFileClip("Game 1 Left.mp4")#.subclip(800,810)  # Left side
video2 = VideoFileClip("Game 1 Right.mp4")#.subclip(800,810)#34.4)  # Right side

# === Define overlap in pixels (adjust based on your footage) ===
overlap_px = 130  # Try 80–150 depending on how much the cameras overlap

# === Crop each video to remove overlap ===
left_cropped = video0.crop(x2=video0.w - overlap_px)
right_cropped = video2.crop(x1=overlap_px)

# === Align right video next to left ===
right_positioned = right_cropped.set_position((left_cropped.w, 0))

# === Combine both into one stitched view ===
stitched_width = left_cropped.w + right_cropped.w
stitched_height = max(left_cropped.h, right_cropped.h)

stitched_clip = CompositeVideoClip(
    [left_cropped, right_positioned],
    size=(stitched_width, stitched_height)
)

# === Preview or export ===
#stitched_clip.preview()  # For notebook testing
stitched_clip.write_videofile("Game_1_Full.mp4", codec="libx264", fps=25)
```
